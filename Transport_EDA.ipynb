# EDA on Transport dataset

```python
#importing the relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data
raw_data = pd.read_csv('Mode_of_Transport.csv')
```

```python
# Make a copy of that data, which can be used for further operations.
df = raw_data.copy()
```

```python
# checking the top 5 values
df.head()
```

```python
# checking the bottom 5 values
df.tail()
```

```python
# Checking the shape
df.shape
```

```python
#Getting the total information
df.info()
```

```python
# Graduate column is having one null value.We have to fill it
df.Graduate = df.Graduate.fillna(0)
```

```python
# Checking for any null values
df.Graduate.isna().sum()
```

```python
df.head()
```

```python
df.Gender.value_counts().plot(kind = 'pie',autopct = '%.f%%')
plt.show()
```

We got most of the data from Male 

```python
# Lets see Genderwise usgae of Transport Facility
sns.countplot(x = df['Gender'],hue= df['Transport'])
plt.show()
```

From above we can see that, both Male & Fmale are preferring the public transport over others.

```python
df.Employed.value_counts()
```

```python
df.Employed.value_counts().plot(kind='pie',autopct = '%.f%%')
plt.show()
```

We can see that, 75% of our population posses certain job.

```python
# Lets see about the usage of Taransport based on employability
sns.countplot(x= df["Employed"], hue=df['Transport'])
plt.show()
```

Here also both epmloyed and notemployed preferring Public Transport over others

```python
sns.countplot(x = df['Graduate'],hue = df['Transport'])
plt.show()
```

So here also we are seeing same pattern that majorly Public Transport is prefered irrespective of education.

```python
N= 444
colors = np.random.rand(N)
plt.scatter(x=df['Work Exp'],y = df['Salary'],c = colors)
plt.xlabel('Work Experience',fontdict = {'font':'arial','size':15,'color':'green'})
plt.ylabel('Salary',fontdict = {'font':'arial','size':15,'color':'green'})
plt.title('Salary Vs Experience',fontdict = {'font':'arial','size':20,'color':'magenta'})
plt.show()
```

From above we can conclude that, with increase in experience salary of employees increasing linearaly.

```python
df.head()
```

```python
# Let us see about the usage of Transport w.r.t salary
sns.displot(data= df, x= 'Salary',hue= 'Transport',palette='seismic')
plt.show()
```

From above we can conclude, the person whose salary is more than 30-35 units is more likely to use private transport for travelling.

```python
# Let us see about the usage of Transport w.r.t distance
sns.displot(data= df, x= 'Distance',hue= 'Transport',palette='rocket')
plt.show()
```

```python
# Let us see about the usage of Transport w.r.t distance
sns.distplot(df['Age'])
plt.show()
```

We can see that, for shorter distances mostly prefered transport mode is public only but as travel distance increases beyond 10-12.5 units then the prefernce is given for other or private transport 

```python
# some information about person's education and employability
sns.countplot(x='Employed',hue='Graduate',data=df)
```
